```python
import sys
import pandas as pd
import numpy as np
import sklearn
import keras
```

```python
names = ['n_pregnant', 'glucose_concentration', 'blood_pressuer (mm Hg)', 'skin_thickness (mm)', 'serum_insulin (mu U/ml)',
        'BMI', 'pedigree_function', 'age', 'class']
# import the uci pima indians diabetes dataset
df = pd.read_csv('diabetes.csv', names = names)
```

# Preprocess the data, mark zero values as NaN and drop


```python
#bmi=0, glucose=0=>mising data
df[df['glucose_concentration'] == 0]
```

```python
#for columns like glucose,etc where 0 makes no sense
columns = ['glucose_concentration', 'blood_pressuer (mm Hg)', 'skin_thickness (mm)', 'serum_insulin (mu U/ml)', 'BMI']

for col in columns:
    df[col].replace(0, np.NaN, inplace=True)
    
df.describe()
#doesnot reads nan values
```

```python
# Drop rows with missing values
df.dropna(inplace=True)

# summarize the number of rows and columns in df
df.describe()
```

```python
# Convert dataframe to numpy array
dataset = df.values#only values no axes labels
print(dataset.shape)
```

```python
# split into input (X) and an output (Y)
X = dataset[:,0:8]#everything from first 7 columns
Y = dataset[:, 8].astype(int)#class is a float vlaue
```

```python
print(X.shape)
print(Y.shape)
print(X[:5])
print(Y[:5])
#0=>healthy,1=>diabetes
```

```python
# Standardize the data using sklearn StandardScaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X)#didn't standardize the data,just creates the matrix
#becasue some values are much greater than other values so they will be weighted more heavily by algorithm just based on size
```

```python
print(scaler)
```

```python
X_standardized=scaler.transform(X)

data=pd.DataFrame(X_standardized)
data.describe()
```

```python
#defining the model(in keras)
def create_model():
    model=Sequential()#using sequential model
    #first layer is the input layer(already defined)
    
    model.add(Dense(8, input_dim = 8, kernel_initializer='normal', activation='relu'))
    model.add(Dense(4, input_dim = 8, kernel_initializer='normal', activation='relu'))
    
    #output layer
    model.add(Dense(1, activation='sigmoid'))
    
    # compile the model
    adam = Adam(lr = 0.01)#learning rate
    model.compile(loss = 'binary_crossentropy', optimizer = adam, metrics = ['accuracy'])
    #whether we getting correct results or not
    
    return model

model = create_model()
print(model.summary())
```

# Do a grid search for the optimal batch size and number of epochs

```python
seed=6
np.random.seed(seed)
#helping to create eproducable results; we have random initialization of weights; 
#we have random divisions of data into groups
#but still here might have got different reuslts cuz there is more to reproducibilty when grid searching wrapped keras models then presented in this tutorial

def create_model():
    model=Sequential()
    #first layer is the input layer(already defined)
    
    model.add(Dense(8, input_dim = 8, kernel_initializer='normal', activation='relu'))
    model.add(Dense(4, input_dim = 8, kernel_initializer='normal', activation='relu'))
    
    #output layer
    model.add(Dense(1, activation='sigmoid'))
    
    # compile the model
    adam = Adam(lr = 0.01)#learning rate
    model.compile(loss = 'binary_crossentropy', optimizer = adam, metrics = ['accuracy'])
    return model

#create the model
model=KerasClassifier(build_fn=create_model,verbose=1)# vrbose=0 print nothing

#define the grid search parameters (so it knows what its going to optimize)
batch_size=[10,20,40] #no. of steps algo should take /no. of inputs algo should look at before
#before it sums all the changes to the gradients and the updates the weight parameters

epochs=[10,50,100]#how long we train network for

#make a dictionary of grid search parameters(required for gridsearchcv)
param_grid = dict(batch_size=batch_size, epochs=epochs)

# build and fit the GridSearchCV 
#here ww will be doing hyperparameter tuning
grid = GridSearchCV(estimator = model, param_grid = param_grid, cv = KFold(random_state=seed), verbose = 10)
#cv tests on every data point cuz maybe some data points are harder to classify then others(k=3 default,5 or 10 more stable)
grid_results = grid.fit(X_standardized, Y)

#summarize the results
print("Best: {0}, using {1}".format(grid_results.best_score_, grid_results.best_params_))
means = grid_results.cv_results_['mean_test_score']
stds = grid_results.cv_results_['std_test_score']
params = grid_results.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print('{0} ({1}) with: {2}'.format(mean, stdev, param))

#found out that during kfold cross validation, accuracy on traing dataset was much higher than on cross validation set;prompting overfitting
# %age of accuracy matters;need to chose the best;need to be sure when classifying as diabetic or not
```

# Do a grid search for learning rate and dropout rate


```python
# import necessary packages
from keras.layers import Dropout

# Define a random seed
seed = 6
np.random.seed(seed)

# Start defining the model
def create_model(learn_rate, dropout_rate):
    # create model
    model = Sequential()
    model.add(Dense(8, input_dim = 8, kernel_initializer='normal', activation='relu'))
    model.add(Dropout(dropout_rate))#periodically droputs neaurons;
    #preventing any specific neuron to be heavily weighted int he whole network=>reducing oerfitting and generalizing better to new data
    #dropout rate: 0(no dropout) to 1(all neurons dropped)
    
    model.add(Dense(4, input_dim = 8, kernel_initializer='normal', activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1, activation='sigmoid'))
    
    # compile the model
    adam = Adam(lr = learn_rate)
    model.compile(loss = 'binary_crossentropy', optimizer = adam, metrics = ['accuracy'])
    return model

# create the model
model = KerasClassifier(build_fn = create_model, epochs = 100, batch_size = 20, verbose = 0)
#setting batch size and epochs to best values from above code

# define the grid search parameters
learn_rate = [0.001, 0.01, 0.1]
dropout_rate = [0.0, 0.1, 0.2]

# make a dictionary of the grid search parameters
param_grid = dict(learn_rate=learn_rate, dropout_rate=dropout_rate)

# build and fit the GridSearchCV
grid = GridSearchCV(estimator = model, param_grid = param_grid, cv = KFold(random_state=seed), verbose = 10)
grid_results = grid.fit(X_standardized, Y)

# summarize the results
print("Best: {0}, using {1}".format(grid_results.best_score_, grid_results.best_params_))
means = grid_results.cv_results_['mean_test_score']
stds = grid_results.cv_results_['std_test_score']
params = grid_results.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print('{0} ({1}) with: {2}'.format(mean, stdev, param))
```

# Do a grid search to optimize kernel initialization and activation functions


```python
# Define a random seed
seed = 6
np.random.seed(seed)

# Start defining the model
def create_model(activation, init):
    # create model
    model = Sequential()
    model.add(Dense(8, input_dim = 8, kernel_initializer= init, activation= activation))
    #model.add(Dropout(0.0))#set from above code
    model.add(Dense(4, input_dim = 8, kernel_initializer= init, activation= activation))
    #model.add(Dropout(0.0))    
    model.add(Dense(1, activation='sigmoid'))
    
    # compile the model
    adam = Adam(lr = 0.001)#set from above code
    model.compile(loss = 'binary_crossentropy', optimizer = adam, metrics = ['accuracy'])
    return model

# create the model
model = KerasClassifier(build_fn = create_model, epochs = 100, batch_size = 20, verbose = 0)

# define the grid search parameters
activation = ['softmax', 'relu', 'tanh', 'linear']
init = ['uniform', 'normal', 'zero']

# make a dictionary of the grid search parameters
param_grid = dict(activation = activation, init = init)

# build and fit the GridSearchCV
grid = GridSearchCV(estimator = model, param_grid = param_grid, cv = KFold(random_state=seed), verbose = 10)
grid_results = grid.fit(X_standardized, Y)

# summarize the results
print("Best: {0}, using {1}".format(grid_results.best_score_, grid_results.best_params_))
means = grid_results.cv_results_['mean_test_score']
stds = grid_results.cv_results_['std_test_score']
params = grid_results.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print('{0} ({1}) with: {2}'.format(mean, stdev, param))
```

# Do a grid search to find the optimal number of neurons in each hidden layer

```python
seed = 6
np.random.seed(seed)

# Start defining the model
def create_model(neuron1, neuron2):
    # create model
    model = Sequential()
    model.add(Dense(neuron1, input_dim = 8, kernel_initializer= 'normal', activation= 'tanh'))
    model.add(Dense(neuron2, input_dim = neuron1, kernel_initializer= 'normal', activation= 'tanh'))
    model.add(Dense(1, activation='sigmoid'))#cuz binary classification problem
    
    # compile the model
    adam = Adam(lr = 0.001)
    model.compile(loss = 'binary_crossentropy', optimizer = adam, metrics = ['accuracy'])
    return model

# create the model
model = KerasClassifier(build_fn = create_model, epochs = 100, batch_size = 20, verbose = 0)

# define the grid search parameters
neuron1 = [4, 8, 16]
neuron2 = [2, 4, 8]
#neural network size(4,2)<(16,8)
#say,if diabetes is highly dependent on bmi and other smaller set of attributes, small neural network is better
# match the complexity of neural network with cmplexity of underlying data
#suggestion:start with a coarse view of network and then make it finer
     #say number of neurons=1,10,50,100 and then fine tune to say 15,16,17

    # make a dictionary of the grid search parameters
param_grid = dict(neuron1 = neuron1, neuron2 = neuron2)

# build and fit the GridSearchCV
grid = GridSearchCV(estimator = model, param_grid = param_grid, cv = KFold(random_state=seed), refit = True, verbose = 10)
grid_results = grid.fit(X_standardized, Y)

# summarize the results
print("Best: {0}, using {1}".format(grid_results.best_score_, grid_results.best_params_))
means = grid_results.cv_results_['mean_test_score']
stds = grid_results.cv_results_['std_test_score']
params = grid_results.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print('{0} ({1}) with: {2}'.format(mean, stdev, param))
```
